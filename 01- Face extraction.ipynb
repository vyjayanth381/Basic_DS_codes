```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading image 
img = cv2.imread('pic.png') # read in gray scale
plt.imshow(img)
```

```python
# convert into grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
faces= cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
position = faces.detectMultiScale(gray,1.3,5)
position
```

```python
for x,y,w,h in position:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),1)
    
plt.figure(figsize=(6,6))
plt.imshow(img)
```

```python
# crop
crop = gray[y:y+h,x:x+w]
plt.imshow(crop,cmap='gray')
```

```python
# resize
img_re = cv2.resize(crop,(64,64))
plt.imshow(img_re)
```

```python
cv2.imwrite('male.png',img_re)
```
