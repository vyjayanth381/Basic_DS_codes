```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
df= pd.read_csv('F:Days Storage.csv')
df.head()
```

```python
sns.scatterplot('Days','GM',data=df)
#scatterplot or relplot
```

```python
df['Days^2']= df.Days**2
```

```python
df1=df[['Days','GM','Days^2']]
df1.head()
```

```python
df1['Days^3']=df1.Days**3
df1['log(Days)']= np.log(df1['Days'])
```

```python
sns.relplot('Days','GM',data=df1,aspect=2)
```

```python
sns.relplot('Days^2','GM',data=df1,aspect=2)
```

```python
sns.relplot('Days^3','GM',data=df1,aspect=2)
```

```python
sns.relplot('log(Days)','GM',data=df1,aspect=2)
```

```python
#SUBPLOT:
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
sns.scatterplot(x=df1['Days'],y=df1['GM']),plt.grid()
plt.subplot(2,2,2)
sns.scatterplot(x=df1['Days^2'],y=df1['GM']),plt.grid()
plt.subplot(2,2,3)
sns.scatterplot(x=df1['Days^3'],y=df1['GM']),plt.grid()
plt.subplot(2,2,4)
sns.scatterplot(x=df1['log(Days)'],y=df1['GM']),plt.grid()
plt.show()
```
