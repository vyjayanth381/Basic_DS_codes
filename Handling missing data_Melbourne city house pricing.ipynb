### treating missing values:
- there are two wasy to treat missing values
1. delete: Delete missing values
2. impute:
    - Imputing by a simple satistic: Replace the missing values by another value, commonly the mean, median, mode
    - Predictive techniques: Use statistical models such as K-NN, SVM etc., to predict and impute missing values.
- In generally, Imputation makes assumption about the missing values and replace missing values by arbitary numbers such as mean, median, mode etc,.. it should be used only when you are reasonably confident about the assumptions
- Otherwise, deletion is often safes and recommended. You may lose some data, but will not make any unreasonable assumptions.

Cautions: Always have a backup of the original data if you're deleting missing values

```python
# removing NaN price in rows
df= df[~np.isnan(df['Price'])]

round(100*(df.isnull().sum()/len(df.index)),2)
```
