#### Find the number of married males that does not buy the bike:

```python
data1['Marital Status'][(data1['Gender']=='Male') & (data1['Bike Buyer']=='No')].value_counts()
```

**OR**

```python
data1['Marital Status'][(data1['Gender']=='Male')&(data1['Bike Buyer']=='No')&(data1['Marital Status']=='Married')].value_counts()
```

#### Find the number of Partial College Home Owners who buy the bike:

```python
(data1['Education']=='Partial College')[(data1['Home Owner']=='Yes')&(data1['Bike Buyer']=='Yes')].value_counts()
```

**OR**

```python
data1['Education'][(data1['Home Owner']=='Yes')&(data1['Bike Buyer']=='Yes')&(data1['Education']=='Partial College')].value_counts()
```

#### Find the number of unmarried Professional Employees that does not buy a bike:

```python
(data1['Marital Status']=='Single')[(data1['Occupation']=='Professional')&(data1['Bike Buyer']=='No')].value_counts()
```

```python
data1['Marital Status'][(data1['Occupation']=='Professional')&(data1['Bike Buyer']=='No')&(data1['Marital Status']=='Single')].value_counts()
```

#### Find the number of Bike Buyers where the person is from High Scool with a Yearly Income less than 62500

```python
(data1['Education']=='High School')[(data1['Yearly Income']<62500)&(data1['Bike Buyer']=='Yes')].value_counts()
```

**OR**

```python
data1['Yearly Income'][(data1['Education']=='High School')&(data1['Bike Buyer']=='Yes')&(data1['Yearly Income']<62500)].value_counts()
```
