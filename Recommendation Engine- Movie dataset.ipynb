#### Calculating Cosine Similarity between Users:

```python
from sklearn.metrics import pairwise_distances
from sklearn.metrics import euclidean_distances
from sklearn.metrics import jaccard_similarity_score
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1- pairwise_distances(user_movie_df.values,metric='cosine')# default metric is euclidean dist

# store the results in a dataframe
user_sim_df_cos = pd.DataFrame(user_sim)
```

```python
user_sim_df_cos.head()
```

```python
user_sim_df_cos.shape
```

```python
user_sim_df_cos.index = df['userId'].unique()
user_sim_df_cos.columns = df['userId'].unique()
```

```python
user_sim_df_cos.head()
```

```python
#fill diagonal with zero 
np.fill_diagonal(user_sim,0)
user_sim_df_cos.head()
```

### Finding common movies of similar Users:

```python
movies_df =pd.read_csv('F://movies.csv')
```

```python
movies_df.head()
```

```python
movies_df.drop('genres',axis=1,inplace=True)
movies_df.head()
```

![img](https://i.postimg.cc/fwG782ps/joins.png)

```python
def get_user_similar_movies(user1,user2):
    # Inner join b/w movies watched b/w 2 users will give
    # the common movies watched.
    common_movies =  df[df['userId']== user1].merge(
        df[df['userId']== user2], on = 'movieId', how = 'inner')
    
    return common_movies.merge(movies_df, on= 'movieId')
```

```python
common_movies =  get_user_similar_movies(1,266)
```

```python
common_movies
```

```python
# not common movies in user 1 and 266 :
# from that recommend movies to user1.
# check what movie are not common movies in 266
```

```python
def get_user_not_similar_movies(user1,user2):
    # Inner join b/w movies watched b/w 2 users will give
    # the common movies watched.
    not_common_movies =  df[df['userId']== user1].merge(df[df['userId']== user2], on = 'movieId', how = 'outer')
    
    return not_common_movies.merge(movies_df, on= 'movieId')
```

```python
not_common_movies=  get_user_not_similar_movies(1,266)
```
